# 13Page 결측치 제거

```r
# 결측치 확인
summary(titanic_train)
```

```r
# 결측치 처리 방법
# 1. 제거
titanic_train_no = na.omit(titanic_train)

# 2. 평균값으로 대체
titanic_train_mean = titanic_train
titanic_train_mean$Age[is.na(titanic_train_mean$Age)] = mean(titanic_train_mean$Age,na.rm = T)

# 3. 특정 조건의 평균값으로 대체
titanic_train_sx = titanic_train
titanic_train_sx$Age[is.na(titanic_train_sx$Age) & titanic_train_sx$Sex == 'female'] = mean(titanic_train_sx$Age[titanic_train_sx$Sex == 'female'],na.rm = T)
titanic_train_sx$Age[is.na(titanic_train_sx$Age) & titanic_train_sx$Sex == 'male'] = mean(titanic_train_sx$Age[titanic_train_sx$Sex == 'male'],na.rm = T)

# 4. Age를 예측할ㅇ 수 있는 알고리즘을 활용하여 Age를 예측 후 대체
# 추후에 진행

print(mean(titanic_train_no$Age)) # 데이터 값 확인
print(mean(titanic_train_mean$Age)) # 데이터 값 확인
print(mean(titanic_train_sx$Age)) # 데이터 값 확인
```

# 18Page 데이터 구간화

```r
# 데이터 구간화
# 타이타닉 데이터에는 진행 X
titanic_train_no$Age_cd = titanic_train_no$Age
titanic_train_no$Age_cd[titanic_train_no$Age < 10] = 'under 10'
titanic_train_no$Age_cd[titanic_train_no$Age >= 10 & titanic_train_no$Age < 20] = '10 ~ 20'
titanic_train_no$Age_cd[titanic_train_no$Age >= 20 & titanic_train_no$Age < 30] = '20 ~ 30'
titanic_train_no$Age_cd[titanic_train_no$Age >= 30 & titanic_train_no$Age < 40] = '30 ~ 40'
titanic_train_no$Age_cd[titanic_train_no$Age >= 40 & titanic_train_no$Age < 50] = '40 ~ 50'
titanic_train_no$Age_cd[titanic_train_no$Age >= 50 & titanic_train_no$Age < 60] = '50 ~ 60'
titanic_train_no$Age_cd[titanic_train_no$Age >= 60] = 'over 60'

titanic_train_no$Age_cd = as.factor(titanic_train_no$Age_cd)
```

# 43Page 실습 정답

```r
# A1 : 의사결정나무 모델을 생성하라
# hint : install.packages(), library(), rpart()
install.packages('rpart')
library(rpart)

dt = rpart(Survived ~ ., train)
```

# 48Page 모델 평가지표 예제

```r
# 모델을 통한 예측값 추출
pred = predict(dt, type = 'class')

head(pred)
```

```r
# 실제 정답 추출
real = train$Survived

# confusion Matrix 생성
table(pred = pred, real = real)
```
